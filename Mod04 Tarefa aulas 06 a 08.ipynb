````python
#1) Seu código aqui
#Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município), 
#e que devolve os dados da consulta no formato DataFrame.

def df_dados(UF:str, tipo:str):
    
    url = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={UF}&tipo_interessado={tipo}'
    r = requests.get(url)
    data_json = r.json()
    dados= pd.DataFrame(data_json['items'])
    dados
    return dados

dados = df_dados(UF = 'SP', tipo = 'Estado')

dados.head(5)
````

````python
# 2) Seu código aqui
#Quantas solicitações para o Estado podem ser consultadas 
#para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas?
                    
df_mg = df_plv[df_plv['uf']=='MG']
df_mg_arq = df_mg[df_mg['status']=='Arquivado por decurso de prazo']
df_mg_arq.value_counts('id_pleito') #quantidade de solicitacoes para os filtros selecionados 70.
````
